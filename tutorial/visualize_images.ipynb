# Notebook to plot image in a grid

## Import

```python
import os
import random

import cv2
import matplotlib
import matplotlib.gridspec as gridspec
from matplotlib import pyplot as plt
```

## Set paths and read images

```python
root_path = str = "D:/DOMI/University/Thesis/Results/plots/images_GT_SR"
test_sets = os.listdir(root_path)
test_set = "Novara_good"
path_lr: str = f"{root_path}/{test_set}/lr"
path_sr_from_lr: str = f"{root_path}/{test_set}/sr1_from_lr"
path_gt: str = f"{root_path}/{test_set}/gt"
path_sr0: str = f"{root_path}/{test_set}/sr0"
path_sr1: str = f"{root_path}/{test_set}/sr1"
path_sr2: str = f"{root_path}/{test_set}/sr2"

path_plot: str = "D:/DOMI/University/Thesis/Results/plots/grid_plots/"+test_set

filenames_img_lr: str = os.listdir(path_lr)
filenames_img_sr_from_lr: str = os.listdir(path_sr_from_lr)
filenames_img_gt: str = os.listdir(path_gt)
filenames_img_sr0: str = os.listdir(path_sr0)
filenames_img_sr1: str = os.listdir(path_sr1)
filenames_img_sr2: str = os.listdir(path_sr2)
```

## Plot

### Matplotlib style settings

```python
SMALL_SIZE = 30
matplotlib.rc('font', size=SMALL_SIZE)
matplotlib.rc('axes', titlesize=SMALL_SIZE)
```

### Plotting

```python
fig = plt.figure(figsize=(20,20))
gs1 = gridspec.GridSpec(4,4)
gs1.update(wspace=0.02, hspace=0.02)
# for i, gt, sr0, sr1, sr2 in zip(range(4), filenames_img_gt, filenames_img_sr0, filenames_img_sr1, filenames_img_sr2):
for i, gt, sr1, lr, sr_lr in zip(range(4), filenames_img_gt, filenames_img_sr1, filenames_img_lr, filenames_img_sr_from_lr):
    image_ID = gt.split(".")[0].split("_")[0]

    image_lr = cv2.imread(os.path.join(path_lr,lr))
    image_lr = cv2.cvtColor(image_lr, cv2.COLOR_BGR2RGB)
    image_sr_lr = cv2.imread(os.path.join(path_sr_from_lr,sr_lr))
    image_sr_lr = cv2.cvtColor(image_sr_lr, cv2.COLOR_BGR2RGB)

    image_gt = cv2.imread(os.path.join(path_gt,gt))
    image_gt = cv2.cvtColor(image_gt, cv2.COLOR_BGR2RGB)


    image_sr1 = cv2.imread(os.path.join(path_sr1, sr1))
    image_sr1 = cv2.cvtColor(image_sr1, cv2.COLOR_BGR2RGB)

    # image_sr0 = cv2.imread(os.path.join(path_sr0, sr0))
    # image_sr0 = cv2.cvtColor(image_sr0, cv2.COLOR_BGR2RGB)
    # image_sr2 = cv2.imread(os.path.join(path_sr2, sr2))
    # image_sr2 = cv2.cvtColor(image_sr2, cv2.COLOR_BGR2RGB)

    # ax1 = plt.subplot(gs1[0,i])
    # ax1.imshow(image_gt), ax1.set_aspect('equal')
    # ax1.set_yticks([]), ax1.set_xticks([])
    # if i == 0:
    #     ax1.set_ylabel("GT")
    # plt.title(f"#{i+1}")

    ax2 = plt.subplot(gs1[1,i])
    ax2.imshow(image_gt), ax2.set_aspect('equal')
    ax2.set_yticks([]), ax2.set_xticks([])
    if i == 0:
        ax2.set_ylabel("GT")
    plt.title(f"#{i+1}")

    ax3 = plt.subplot(gs1[2,i])
    ax3.imshow(image_sr1), ax3.set_aspect('equal')
    ax3.set_yticks([]), ax3.set_xticks([])
    if i == 0:
        # ax3.set_ylabel("SR (V1)")
        ax3.set_ylabel("SR from GT")

    # ax2 = plt.subplot(gs1[1,i])
    # ax2.imshow(image_sr0), ax2.set_aspect('equal')
    # ax2.set_yticks([]), ax2.set_xticks([])
    # if i == 0:
    #     ax2.set_ylabel("SR (V0)")

    # ax4 = plt.subplot(gs1[3,i])
    # ax4.imshow(image_sr2), ax4.set_aspect('equal')
    # ax4.set_yticks([]), ax4.set_xticks([])
    # if i == 0:
    #     ax4.set_ylabel("SR (V2)")
    
    ax1 = plt.subplot(gs1[0,i])
    ax1.imshow(image_lr), ax1.set_aspect('equal')
    ax1.set_yticks([]), ax1.set_xticks([])
    if i == 0:
        ax1.set_ylabel("LR")
    plt.title(f"#{i+1}")

    ax4 = plt.subplot(gs1[3,i])
    ax4.imshow(image_sr_lr), ax4.set_aspect('equal')
    ax4.set_yticks([]), ax4.set_xticks([])
    if i == 0:
        ax4.set_ylabel("SR from LR")
# fig.suptitle(f"Dataset: {test_set}")


plt.savefig(f"{os.path.join(path_plot)}/{test_set}.svg", format="svg")
plt.savefig(f"{os.path.join(path_plot)}/{test_set}.png")
plt.savefig(f"{os.path.join(path_plot)}/{test_set}.jpg")
# plt.savefig(f"{os.path.join(path_plot)}/{test_set}_models.svg", format="svg")
# plt.savefig(f"{os.path.join(path_plot)}/{test_set}_models.png")
# plt.savefig(f"{os.path.join(path_plot)}/{test_set}_models.jpg")
plt.show()
```
